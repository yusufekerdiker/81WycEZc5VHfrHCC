```python
# libraries that i need
import tweepy # for twitter api and search
import pandas as pd # creating csv, csv to html, creating dataframes

# twitter developer authentication keys and tokens
# if the keys are in different folder use ==> %run ./(key-folder-name)
consumer_key="mykeyswashere"
consumer_secret_key="mykeyswashere"
access_token="mykeyswashere"
access_token_secret="mykeyswashere"

# online authentication
auth=tweepy.OAuthHandler(consumer_key, consumer_secret_key)
auth.set_access_token(access_token, access_token_secret)
api=tweepy.API(auth)

num_of_tweets = 25 # Change here depending on how many tweets you want to see

name = []
dates = []
tweets = []
likes = []
retweets = []
discussions = []

query = "request for startup -filter:retweets" # add "AND -filter:replies" inside of query for delete replied tweets

for item in tweepy.Cursor(api.search, q=query, tweet_mode = "extended", truncated=False).items(num_of_tweets):
    name.append(item.user.screen_name) # or item.author.screen_name
    dates.append(item.created_at)
    tweets.append(item.full_text)
    likes.append(item.favorite_count)
    retweets.append(item.retweet_count)
    discussions.append(item.contributors) # I didn't understand neither the discussions nor the contributors on Twitter, 
                                          # but because the company wanted it, I tried to add it as much as possible.
    
rtweets = sorted(retweets, reverse=True) 
# sorts most retweeted tweets(?) and returns at descending order / reverse=True mean descending

mostliked = sorted(likes, reverse=True) 
# returns most liked tweets -- descending order

time = sorted(dates, reverse=True) 
# the date is already in descending order, also if you add "reverse = True" nothing changes, it just allows you to be sure.
# but if you change "reverse=True" to "reverse=False" sorting becomes ascending order

data_table = pd.DataFrame({'Author:':name, 
                           'Date and Time:':dates,
                           'Tweet Content:':tweets, 
                           'Number of Likes:':likes, 
                           'Number of Retweets:':retweets,
                           'Number of Discussions:':discussions})

data_table.index.rename('Index:', inplace=True) # or data_table.index.name = 'Index:'
# inplace means overwrite the existing dataframe
# Added name to unnamed numbering in Pandas' data frame

data_table.reset_index(drop=True) # resets the index error of the data table

data_table.to_csv('twitter-query.csv', encoding='utf-8', header='true', index=True)

data_table
```
